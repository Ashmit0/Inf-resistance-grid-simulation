### Analysis 
#### Importing libraries and data

```python
import matplotlib.pyplot as plt 
import pandas as pd 
data = pd.read_csv( '/Users/ashmitbathla/Desktop/ltspice-nrtlist/inf-R-grid/data.csv' )
print( data.head() )
```

#### Resistance count

The general recurrance relation for resistance count ($f(n)$) is give as (proof by induction); 
$$f(n) = f(n-2) + 4n + 8 $$ 
from order analysis, it is clear that resistance count is of the order $\mathcal{O}(n^2)$ in $n$, where $n$ = lattice length 

```python
plt.plot( data.length , data.resistance_cnt )
plt.grid( True )
plt.xlabel('length of the lattice')
plt.ylabel('total number of resistances')
plt.title('resistance cnt vs lattice length')
plt.show()
```

###  Impedance 

The net impedance of the circuit quickly converges to the inf limit, form the plot it is clear that, 
$$\lim_{l \to \infty} Z = 0.773 \Omega$$

```python
plt.plot( data.length , data.impedance )
plt.grid( True )
plt.xlabel('length of the lattice')
plt.ylabel('net impedance')
plt.title('net impedance vs lattice length')
plt.show()
```
